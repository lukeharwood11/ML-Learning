### `plt.plot()` Method
plt.plot() accepts 3 args: (x, y, format)

**format** is a short hand combination of _{color}{marker}{line}_

### Examples
* 'r*--' : ‘red stars with dashed lines’
* 'ks.' : ‘black squares with dotted line’ (‘k’ stands for black)
* 'bD-.' : ‘blue diamonds with dash-dot line’.

```python
plt.plot([1, 2, 3, 4, 5], [1, 2, 3, 4, 10], 'go')
```

## Drawing two sets of scatterplots in the same plot
just call `plt.plot()` again

```python
plt.plot([1, 2, 3, 4, 5], [1, 2, 3, 4, 10], 'go') # green dots
plt.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 11], 'b*') # blue stars
plt.show()
```

## Adding basic plot features

Title, legend, X and Y axis labels

```python
# Arbitrary data for future use
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 10]
y2 = [2, 3, 4, 5, 11]
```

```python
plt.plot(x, y1, 'go', label='GreenDots')
plt.plot(x, y2, 'b*', label='BlueStars')
plt.title('A Simple ScatterPlot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='best') # legend text comes from the plot's label parameter
plt.show()
```

```python
# With increased size of plot
plt.figure(figsize=(10,7)) # 10 is width, 7 is height
plt.plot(x, y1, 'go', label='GreenDots')
plt.plot(x, y2, 'b*', label='BlueStars')
plt.title('A Simple ScatterPlot')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.legend(loc='best') # legend text comes from the plot's label parameter
plt.show()
```

## `figure`

The `figure` is like a canvas that holds all the subplots and other plot elements inside it.

Subplots within the canvas are called `axes` that are arranged in rows and columns.

Figures have axes which contain [titles, labels, axis]

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4), sharey=True, dpi=120)
```
